### Prepare X

```python
#Standardized data
scaled_x = StandardScaler().fit_transform(raw_x)
df_std_x = pd.DataFrame(data=scaled_x, columns=raw_x.columns)
df_std_x
```

# Test dataset preparation

```python
#Read test dataset
raw_df_test = pd.read_csv('./spam_ham_dataset.csv')
raw_df_test
```

```python
#Reform dataset
df_test = pd.DataFrame(columns=df_std_x.columns)

count = {}
for col in df_std_x.columns:
    count[col] = 0

for data in raw_df_test['text']:
    content = data.split('\r\n')
    for i in range(2, len(content)):
        content[1] = content[1] + ' ' + content[i]
    for k in count.keys():
        count[k] = content[1].count(k)
    df_test = df_test.append(count, ignore_index=True)
df_test
```

```python
#Add prediction column
df_test['Prediction'] = raw_df_test['label_num']
df_test
```

```python
#Create new csv file
df_test.to_csv("Test_dataset.csv", index=False)
```
